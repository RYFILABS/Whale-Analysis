```python
# Initial Imports

import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

%matplotlib inline
```

```python
# Set the file paths and clean data

whale_returns = Path(r"C:\Users\Ryan\Desktop\UM Fintech Bootcamp\Whale-Analysis\Resources\whale_returns.csv")
whale_returns_df = pd.read_csv(whale_returns, index_col="Date", infer_datetime_format=True, parse_dates=True)

algo_returns = Path(r"C:\Users\Ryan\Desktop\UM Fintech Bootcamp\Whale-Analysis\Resources\algo_returns.csv")
algo_returns_df = pd.read_csv(algo_returns, index_col= "Date", infer_datetime_format=True, parse_dates=True)

sp500_history = Path(r"C:\Users\Ryan\Desktop\UM Fintech Bootcamp\Whale-Analysis\Resources\sp500_history.csv")
sp500_history_df = pd.read_csv(sp500_history, index_col="Date", infer_datetime_format=True, parse_dates=True)

sp500_history_df.head()
```

```python
# Whale Returns

# Read the Whale Portfolio daily returns and clean the data

whale_returns_clean = whale_returns_df.dropna()
```

```python
whale_returns_clean.drop_duplicates()
whale_returns_clean.head()
```

```python
# Find any null values

whale_returns_df.isnull().sum()
```

```python
# Drop nulls

whale_returns_clean = whale_returns_df.dropna()
whale_returns_clean.isnull().sum()
```

```python
# Agorithmic Daily Returns

# Read the algorithmic daily returns and clean the data

# Drop Nulls
algo_returns_clean = algo_returns_df.dropna()
```

```python
# Count Nulls
algo_returns_clean.isnull().sum()
```

```python
# S&P 500 Returns

# Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data.


# Reading S&P 500 Closing Prices
sp500_history_df.head()
```

```python
sp500_history_df.dtypes
```

```python
# Fix Data Types

sp500_history_df['Close'] = sp500_history_df['Close'].str.replace('$','')
```

```python
# Calculate Daily Returns

daily_returns_sp500 = sp500_history_df.astype(float).pct_change()

daily_returns_sp500.sort_index()

daily_returns_sp500.head()
```

```python
# Drop null values
daily_returns_sp500.dropna()
daily_returns_sp500.head()
```

```python
# Rename Column
daily_returns_sp500 = daily_returns_sp500.rename(columns={
    "Close": "S&P 500",})
daily_returns_sp500.head()
```

```python
# Combine Whale, Algorithmic, and S&P 500 Returns

combined_df = pd.concat([whale_returns_clean, algo_returns_clean, daily_returns_sp500], axis="columns", join="inner")
combined_df.head()
```

```python
# Calculate and Plot the daily returns and cumulative returns.

# Plot daily returns

combined_df.plot()
```

```python
# Plot cumulative returns

cumulative_returns = (1 + combined_df).cumprod()
cumulative_returns.plot()
print(cumulative_returns.tail())
```
